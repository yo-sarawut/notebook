```python
import joblib

def test_model_list(txt):
    
    model_main = joblib.load('data/model/model_section.pkl')
    products = txt  # Video game consoles : 95
    p = model_main.predict([products])[0] 
    
    if p <1000:
        a = str(p)
        a = ("0"+a)
    else:
        a = str(p)
   
    model_2 = joblib.load('data/model/model_'+a+'.pkl')
    predicts = model_2.predict([products])[0]          
        
    return predicts
```

```python
urls = "data/test-data.xlsx"
sheets = "wto"
data = pd.read_excel(urls, sheet_name=sheets)
data = data[data['level'] == 4]
data
```

```python
data= data.sample(1000)
data
```

```python
df = data[['target', 'description']]
df['description'] = df['description'].str.lower()
df['target'] = df['target'].apply(str)
df
```

```python
df.columns = ['target', 'data']
df
```

```python
df.loc[:, lambda df: ['data']]
```

```python
results = ""
#df.dropna(inplace=True)
data_test = df.loc[:, lambda df: ['data']]
data_targets = df.loc[:, lambda df: ['target']]
d = []
t = []
r = []
c = []
for i in range(len(data_test)):
    results = test_model_list(data_test[i])
    d.append(data_test[i])
    t.append(data_targets[i])
    r.append(results)
    if results ==data_targets[i]:
        c.append(1)
    else:
        c.append(0)

datas = [d,t,r,c]
df_result = pd.DataFrame(datas)
df_result
```

```python
zippedList =  list(zip(d,t,r,c))
df = pd.DataFrame(zippedList, columns = ['data' , 'target', 'result','check']) 
df
```
